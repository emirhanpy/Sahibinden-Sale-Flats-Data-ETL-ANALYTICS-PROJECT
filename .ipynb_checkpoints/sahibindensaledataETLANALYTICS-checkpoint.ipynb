```python
import pandas as pd
import numpy as np
import sqlite3
import glob
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt

log_file = 'log_file.txt'
csv_file = 'transformed_data.csv'

conn = sqlite3.connect('sahibinden_sale_house.db')
table_name = 'SAHIBINDEN_SALE_HOUSE'
```

```python
def extract_from_csv(file_to_process):
    """Reads a csvfile and returns it as a pandas DataFrame."""
    dataframe = pd.read_csv(file_to_process)
    return dataframe

def extract():
    """Extracts a CSV file and returns it as a pandas DataFrame."""
    extracted_data = pd.DataFrame(columns=['title','area','numberOfRooms','price','town','district'])
    for csvfile in glob.glob('*.csv'):
        extracted_data = pd.concat([extracted_data, pd.DataFrame(extract_from_csv(csvfile))], ignore_index=True)
        
    return extracted_data
```

```python
def transform(data):
    """It does only the necessary transformation on the data."""
    del data['Unnamed: 0']

    data['town'] = data['town'].astype(str)
    data['district'] = data['district'].astype(str)
    data['numberOfRooms'] = data['numberOfRooms'].astype(str)
    data = data[~data['price'].str.contains('[^\d.]', regex=True)]
    data['price'] = data['price'].str.replace('.', '').astype(np.int64)
    try:
        data['area'] = data['area'].astype(int)
    except ValueError:
        print('Value Error')
        
    return data
```

```python
def load_data(csv_file, table_name, conn, transformed_data):
    """Loads the data to csv format and db format."""
    transformed_data.to_csv(csv_file)
    transformed_data.to_sql(table_name, conn, if_exists = 'replace', index = False)
```

```python
def log_progress(message):
    """Logs the progress message along with a timestamp to a log file."""
    timestamp_format = '%Y-%h-%d-%H:%M:%S'
    now = datetime.now()
    timestamp = now.strftime(timestamp_format)
    with open(log_file, "a") as f:
        f.write(timestamp + ' ' + message + '\n')
```

```python
log_progress("ETL JOB STARTED.")

log_progress("EXTRACT PHASE STARTED")
extracted_data = extract()
log_progress("EXTRACT PHASE ENDED")

log_progress("TRANSFORM PHASE STARTED")
transformed_data = transform(extracted_data)
log_progress("TRANSFORM PHASE ENDED")

log_progress("LOAD PHASE STARTED")
load_data(csv_file, table_name, conn, transformed_data)
log_progress("LOAD PHASE ENDED")

log_progress("ETL JOB ENDED")
```

```python
"""This query shows us Towns with avg price of Towns"""
query1 = """
SELECT town as TOWN,avg(price) as AVG_PRICE from SAHIBINDEN_SALE_HOUSE
GROUP BY TOWN
ORDER BY AVG_PRICE desc
"""
query1_output = pd.read_sql(query1, conn)
query1_output['AVG_PRICE'] = query1_output['AVG_PRICE'].apply(lambda x: '{:.0F}'.format(x))
query1_output['AVG_PRICE'] = pd.to_numeric(query1_output['AVG_PRICE']).astype(np.int64)/10000000

town_price_data = pd.DataFrame(query1_output)
town_price_data = town_price_data.head(20)
town_price_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.barplot(x='TOWN', y='AVG_PRICE', data=town_price_data, palette='rocket')
plt.xlabel('Town', fontsize=12, color='black')
plt.ylabel('Avg Sale Price (Million)', fontsize=12, color='black')
plt.title('The 20 Most Expensive Towns in Istanbul (2022) ', fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
"""This query shows us Districts with avg price of Districts"""
query2 = """
SELECT district as DISTRICT,avg(price) as AVG_PRICE from SAHIBINDEN_SALE_HOUSE
GROUP BY DISTRICT
ORDER BY AVG_PRICE desc
"""
query2_output = pd.read_sql(query2, conn)
query2_output['AVG_PRICE'] = query2_output['AVG_PRICE'].apply(lambda x: '{:.0F}'.format(x))
query2_output['AVG_PRICE'] = pd.to_numeric(query2_output['AVG_PRICE']).astype(np.int64)/10000000


district_price_data = pd.DataFrame(query2_output)
district_price_data = district_price_data.head(20)
district_price_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.barplot(x='DISTRICT', y='AVG_PRICE', data=district_price_data, palette='flare_r')
plt.xlabel('District', fontsize=12, color='black')
plt.ylabel('Avg Sale Price (Million)', fontsize=12, color='black')
plt.title('The 20 Most Expensive Districts in Istanbul (2022) ', fontsize=20)
plt.xticks(rotation=45)
plt.show()
```

```python
"""This query shows us Number of Rooms with count of Number of Rooms."""
query3 = """
SELECT numberOfRooms as Number_of_Rooms , count(numberOfRooms) as COUNT from SAHIBINDEN_SALE_HOUSE
GROUP BY Number_of_Rooms
ORDER BY COUNT desc
"""
query3_output = pd.read_sql(query3, conn)

count_numberOfRooms_data = pd.DataFrame(query3_output)
count_numberOfRooms_data = count_numberOfRooms_data.head(5)
count_numberOfRooms_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.barplot(x='Number_of_Rooms', y='COUNT', data=count_numberOfRooms_data, palette='Blues_r')
plt.xlabel('Number of Rooms', fontsize=12, color='black')
plt.ylabel('COUNT', fontsize=12, color='black')
plt.title('Count of The 5 Most Common Flat Type in Istanbul (2022) ', fontsize=20)
plt.xticks(fontsize=12)
plt.show()
```

```python
"""This query shows us Number of Rooms with avg price of Number of Rooms."""
query4 = """
SELECT numberOfRooms as Number_of_Rooms,avg(price) as AVG_PRICE from SAHIBINDEN_SALE_HOUSE
WHERE numberOfRooms = '2+1' or numberOfRooms = '3+1' or numberOfRooms = '1+1' or numberOfRooms = '4+1' or numberOfRooms = '4+2'
GROUP BY numberOfRooms
ORDER BY AVG_PRICE desc
"""
query4_output = pd.read_sql(query4, conn)
query4_output['AVG_PRICE'] = query4_output['AVG_PRICE'].apply(lambda x: '{:.0F}'.format(x))
query4_output['AVG_PRICE'] = pd.to_numeric(query4_output['AVG_PRICE']).astype(np.int64)/1000000

price_numberOfRooms_data = pd.DataFrame(query4_output)
price_numberOfRooms_data = price_numberOfRooms_data.head(5)
price_numberOfRooms_data
```

```python
plt.figure(figsize=(12, 6))
sns.set_style("darkgrid")
sns.barplot(x='Number_of_Rooms', y='AVG_PRICE', data=price_numberOfRooms_data, palette='Blues_r')
plt.xlabel('Number of Rooms', fontsize=12, color='black')
plt.ylabel('AVG SALE PRICE (Million)', fontsize=12, color='black')
plt.title('Average Sale Price of The 5 Most Common Flat Type in Istanbul (2022) ', fontsize=18)
plt.xticks(fontsize=12)
plt.show()
```
